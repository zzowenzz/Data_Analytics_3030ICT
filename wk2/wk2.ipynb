## Data Exploration
Data Exploration is the process of analyzing and summarizing datasets to understand their structure, patterns, and relationships before performing further analysis or modeling.

### categorical to quantitative

```python
# Creating a sample dataset with a categorical column
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Chicago', 'New York']
}

# Converting to a DataFrame
df = pd.DataFrame(data)

print(f"Original DataFrame:\n{df}\n")

# Encoding categorical data
df['Gender_Numeric'] = df['Gender'].map({'Female': 0, 'Male': 1})  # Label Encoding

# One-Hot Encoding for City column
df = pd.get_dummies(df, columns=['City'], drop_first=True)  # Convert categorical 'City' into multiple binary columns

# Display the transformed DataFrame
print(f"After converting categorical data to numeric:\n{df}\n")
```

### categorical to categorical

```python
import pandas as pd

# Sample dataset with categorical values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Chicago', 'New York']
}

# Convert to DataFrame
df = pd.DataFrame(data)

print(f"Original DataFrame:\n{df}\n")

# Reclassifying cities into broader categories
city_mapping = {
    'New York': 'Big City',
    'Los Angeles': 'Big City',
    'Chicago': 'Medium City'
}
df['City_Category'] = df['City'].map(city_mapping)

# Mapping gender into a different categorical representation
gender_mapping = {
    'Female': 'Woman',
    'Male': 'Man'
}
df['Gender_Category'] = df['Gender'].map(gender_mapping)


print(f"After reclassifying categorical data:\n{df}\n")
```

### quantitative to quantitative

```python
# Creating a sample dataset with quantitative values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 40, 35, 50, 30],
    'Salary': [50000, 80000, 70000, 120000, 60000]
}

# Converting to a DataFrame
df = pd.DataFrame(data)
print(F"Original DataFrame:\n{df}\n")

# Quantitative to Quantitative Transformation
# 1. Normalization (Scaling values between 0 and 1)
df['Salary_Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

# 2. Binning (Converting continuous data into discrete groups)
bins = [20, 30, 40, 50, 60]  # Age ranges
labels = ['20s', '30s', '40s', '50s']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


print(f"After transforming quantitative data:\n{df}\n")
```

### Linear and non-linear in quantitative to quantitative
Linear Relationship: The equation `y=2x+3` represents a linear relationship, where y increases at a constant rate as x increases. A linear relationship means that the rate of change is constant, and plotting the data results in a straight-line pattern.

Non-linear relationship: The equation `y=x^2` represents a non-linear relationship, where y increases at an increasing rate as x increases. A non-linear relationship means that the rate of change is not constant, and plotting the data results in a curved pattern.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generating data for a linear relationship (y = 2x + 3)
x_linear = np.linspace(0, 10, 20)
y_linear = 2 * x_linear + 3

# Generating data for a non-linear relationship (y = x^2)
x_nonlinear = np.linspace(-5, 5, 20)
y_nonlinear = x_nonlinear ** 2

# Creating a figure with two subplots
plt.figure(figsize=(12, 5))

# Plotting the Linear Relationship
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='blue', label='Data Points')
plt.plot(x_linear, y_linear, color='black', linestyle='-', linewidth=2, label='Linear Fit')  # Line to show trend
plt.title("Linear Relationship (y = 2x + 3)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()

# Plotting the Non-Linear Relationship
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='red', label='Data Points')
plt.plot(x_nonlinear, y_nonlinear, color='black', linestyle='-', linewidth=2, label='Non-Linear Fit')  # Curve to show trend
plt.title("Non-Linear Relationship (y = x^2)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()

# Display the plots
plt.show()

```

### Multiple variables with correlation matrix
Correlation matrix is a symmetric matrix where element `m_{ij}` is the correlation coefficient between the `i`-th and `j`-th variables. It is used to understand the relationship between multiple variables and identify patterns or dependencies in the data. Element on the diagonal is always 1, as it represents the correlation of a variable with itself.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample dataset with meaningful financial data
np.random.seed(42)

data = {
    'Monthly_Income': np.random.randint(3000, 10000, 50),  # Monthly income in dollars
    'Savings': np.random.randint(1000, 7000, 50),  # Savings amount in dollars
    'Monthly_Expenses': np.random.randint(2000, 8000, 50),  # Monthly expenses in dollars
    'Credit_Score': np.random.randint(300, 850, 50)  # Credit score range (300-850)
}

# Converting to DataFrame
df = pd.DataFrame(data)

# Adding correlation: Higher income leads to higher savings, but high expenses reduce savings
df['Savings'] = df['Monthly_Income'] * 0.3 - df['Monthly_Expenses'] * 0.2 + np.random.randint(-500, 500, 50)

# Adding correlation: Credit score is influenced by savings and income
df['Credit_Score'] = (df['Savings'] * 0.05 + df['Monthly_Income'] * 0.01 + 500).astype(int)
df['Credit_Score'] = df['Credit_Score'].clip(300, 850)  # Ensuring values stay within credit score range

# Computing correlation matrix
correlation_matrix = df.corr()

# Plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Financial Data")
plt.show()
```

Monthly income with Savings is strong positive correlation as the value is 0.77, which means that as monthly income increases, savings also increase. On the other hand, Monthly Expenses with Savings is a negative correlation as the value is -0.58, which means that as monthly expenses increase, savings decrease.

## Data Visualization
Data visualization helps in understanding trends, relationships, and patterns in data. Python provides various libraries like Matplotlib and Seaborn for creating effective visualizations.

### Some 2D plots

```python
# Creating sample data for different 2D plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # Sine wave
y2 = np.exp(0.3 * x)  # Exponential function
y3 = np.log(x + 1)  # Logarithmic function

# Creating a figure with three different 2D plots
plt.figure(figsize=(12, 8))

# Line Plot (Sine Wave)
plt.subplot(3, 1, 1)
plt.plot(x, y1, color='blue', linestyle='-', linewidth=2)
plt.title("Sine Wave")
plt.xlabel("X values")
plt.ylabel("sin(x)")

# Scatter Plot (Exponential Growth)
plt.subplot(3, 1, 2)
plt.scatter(x, y2, color='red')
plt.title("Exponential Growth")
plt.xlabel("X values")
plt.ylabel("exp(0.3x)")

# Bar Plot (Logarithmic Function)
plt.subplot(3, 1, 3)
plt.bar(x[::10], y3[::10], color='green', width=0.5)
plt.title("Logarithmic Function")
plt.xlabel("X values")
plt.ylabel("log(x+1)")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

## Dimensionality Reduction
Dimensionality Reduction is the process of reducing the number of features (variables) in a dataset while retaining as much meaningful information as possible. It is commonly used in machine learning, data visualization, and exploratory data analysis to handle high-dimensional data efficiently.

Why do it?
- Reduces Computational Cost  Fewer features mean faster processing.
- Avoids the Curse of Dimensionality  High-dimensional data can make models less efficient.
- Improves Model Performance  Eliminates irrelevant or redundant features, reducing overfitting.
- Enhances Visualization  Helps in plotting high-dimensional data in 2D or 3D for better interpretation.
- Removes Noise  Extracts the most significant components and removes less informative ones.

Principal Component Analysis (PCA)
Principal Component Analysis (PCA) is a popular dimensionality reduction technique used in machine learning, statistics, and data visualization. It transforms high-dimensional data into a lower-dimensional space while retaining the most important information.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Generating 3D data
np.random.seed(42)
x_3d = np.random.rand(200) * 10
y_3d = np.random.rand(200) * 10
z_3d = 2 * x_3d + 3 * y_3d + np.random.randn(200) * 2  # Adding some noise

# Visualizing 3D Data
fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x_3d, y_3d, z_3d, c=z_3d, cmap='viridis', alpha=0.7)
ax1.set_title("3D Data")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")

# Reducing 3D to 2D using PCA
data_3d = np.column_stack((x_3d, y_3d, z_3d)) # stacking three separate 1D arrays into a 2D array
print(x_3d.shape, y_3d.shape, z_3d.shape, data_3d.shape)
pca_2d = PCA(n_components=2) # applyng PCA that identifies the principal components, which finds two new dimensions that capure the most variance from the 3D space
data_2d = pca_2d.fit_transform(data_3d)

# Visualizing Reduced 2D Data
ax2 = fig.add_subplot(122)
ax2.scatter(data_2d[:, 0], data_2d[:, 1], c=z_3d, cmap='viridis', alpha=0.7)
ax2.set_title("Reduced 2D Data (via PCA)")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")

plt.tight_layout()
plt.show()
```
