```python
#TODO: fix the data type of column Date
# df["Date"] = df["Date"].replace(...)
# df["Date"] = pd....(df["Date"])
df["Date"] = pd.to_datetime(df["Date"])
print(f"Data type of Date: {df['Date'].dtypes}") # now it should be datetime64 and we can filter by date
print(f"The earliest date in the dataset: {df['Date'].min()}. The latest date in the dataset: {df['Date'].max()}")
before_10_03 = df[df["Date"] < "2018-10-03"]
print(f"Number of rows before 10th March 2018: {before_10_03.shape[0]}")
print(f"Number of rows after 10th March 2018: {df.shape[0] - before_10_03.shape[0]}")
```
